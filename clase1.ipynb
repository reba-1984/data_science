```python
import numpy as np
import pandas as pd
```

```python
#Creo array de numpy
array_numeros = np.array(range(1, 11, 1))
array_numeros
```

```python
#creo serie de pandas
serie_numeros = pd.Series(range(1, 11, 1))
serie_numeros

```

```python
#Creo array de 2 dimensiones con numpy de 8 x 8 
#Uso arenge() en vez de arrya(range())
ajedrez = np.arange(1, 65, 1).reshape(8,8)
ajedrez
```

```python
ajedrez_df = pd.DataFrame(ajedrez)
print(ajedrez_df)
```

```python
ajedrez_df_modificado = pd.DataFrame(
                        ajedrez,
                        columns= range(1,9),
                        index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
ajedrez_df_modificado
```
